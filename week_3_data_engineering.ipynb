#Exercise 5: Create an inner function to calculate the addition in the following way

```python
def outer (a,b):
  def inner(a,b):
    return a+b
  c=inner(a,b)
  c+=5
  return c

outer(1,2)
```

#Exercise 6: Create a recursive function

```python
def sum(a):
  if a==0:
    return 0
  return a+sum(a-1)

sum(10)
```
